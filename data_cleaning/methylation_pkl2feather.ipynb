# Data cleaning for public methylation data
* Separate metadata and DNA methylation in the pickle file
* Store methylation data as binary matrix format that can be read by R.


```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
train_folder='../extdata/methylation/train_old/'
test_folder='../extdata/methylation/test_old/'
```

```python
train_files = os.listdir(train_folder)
train_files = [tfile for tfile in train_files if tfile.endswith("pkl")]
test_files = os.listdir(test_folder)
test_files = [tfile for tfile in test_files if tfile.endswith("pkl")]
```

```python
for train_file in train_files:
    print(train_file)
    fname = train_file.split(".")[0]
    
    # load pickle file
    with open(os.path.join(train_folder, train_file), 'rb') as f:
        data=pickle.load(f)
    
    # split into metadata and methylation data
    metadata = data.loc[:, ["dataset", "tissue_type", "age", "gender"]]
    methylation_data = data.iloc[:, 4:]
    methylation_data.reset_index(inplace=True, names="Sample")

    # store two datasets back to the folder
    meta_fname = f"{fname}.csv"
    methylation_fname = f"{fname}.feather"
    metadata.to_csv(os.path.join(train_folder, meta_fname))
    methylation_data.to_feather(os.path.join(train_folder, methylation_fname))

    
```

```python
for test_file in test_files:
    print(test_file)
    fname = test_file.split(".")[0]
    
    # load pickle file
    with open(os.path.join(test_folder, test_file), 'rb') as f:
        data=pickle.load(f)
    
    # split into metadata and methylation data
    metadata = data.loc[:, ["dataset", "tissue_type", "age", "gender"]]
    methylation_data = data.iloc[:, 4:]
    methylation_data.reset_index(inplace=True, names="Sample")

    # store two datasets back to the folder
    meta_fname = f"{fname}.csv"
    methylation_fname = f"{fname}.feather"
    metadata.to_csv(os.path.join(test_folder, meta_fname))
    methylation_data.to_feather(os.path.join(test_folder, methylation_fname))
```
